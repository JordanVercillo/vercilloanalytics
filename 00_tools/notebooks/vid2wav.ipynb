```python
import os
from moviepy import VideoFileClip

# Your folder path
folder_path = r"C:\Users\jverc\Downloads\00_notebooks\resources\lectures\04_big_data"

# Define your prefix here
file_prefix = "04_big_data_management_" 

def convert_mp4_to_wav(directory):
    if not os.path.exists(directory):
        print(f"Error: The folder '{directory}' was not found.")
        return

    print(f"Scanning: {directory}...\n")
    files_converted = 0
    
    for filename in os.listdir(directory):
        if filename.lower().endswith(".mp4"):
            mp4_path = os.path.join(directory, filename)
            
            # UPDATED: Add prefix to the filename
            new_filename = file_prefix + filename[:-4] + ".mp3"
            output_path = os.path.join(directory, new_filename)
            
            print(f"Converting: {filename} -> {new_filename}...")
            
            try:
                video = VideoFileClip(mp4_path)
                
                if video.audio is not None:
                    # UPDATED: Removed 'verbose' and 'logger' arguments
                    video.audio.write_audiofile(output_path) 
                    
                    files_converted += 1
                    print("   [OK] Done.")
                else:
                    print("   [SKIP] Video has no audio track.")
                
                video.close()
                
            except Exception as e:
                print(f"   [ERROR] Failed: {e}")

    print(f"\nCompleted! Converted {files_converted} files.")

convert_mp4_to_wav(folder_path)
```
