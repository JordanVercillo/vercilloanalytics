```python
from PyPDF2 import PdfReader

reader = PdfReader(file_path)
all_text = ""
for page in reader.pages:
    all_text += page.extract_text() + "\n"

print(all_text)
```

```python
# Split into lines
lines = [line.strip() for line in all_text.splitlines() if line.strip()]
lines
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(lines, columns=["text"])
df
```

```python
pattern1 = r"Your\s*Transactions\s*The\s*American\s*Express\s*Gold\s*Rewards\s*Card"
pattern2 = r"Prepared\s*For\s*Account\s*Number\s*Opening\s*Date\s*Closing\s*Date"

conditions = [
    df["text"].str.contains(pattern1, na=False, regex=True),
    df["text"].str.contains(pattern2, na=False, regex=True)
]

print(df[conditions[0]])
print(df[conditions[1]])
```

```python
# replace pattern 1
df["clean_text"] = df["text"].str.replace(
    pattern1,        # regex pattern to remove
    "",              # replace with empty string
    regex=True
)

# replace pattern 2 on the result
df["clean_text"] = df["clean_text"].str.replace(
    pattern2,
    "",
    regex=True
)
```

```python
df = df["clean_text"]
df.head(10)
```

```python
import re

# Regex for opening and closing dates
date_pattern = re.compile(r"([A-Za-z]{3,9}\s?\d{1,2},\s\d{4})")

# Search all rows
opening_date = None
closing_date = None

for line in df.tolist():
    matches = date_pattern.findall(line)
    if len(matches) == 2:
        opening_date, closing_date = matches
        break  # Stop after first match

print("Opening Date:", opening_date)
print("Closing Date:", closing_date)
```

```python
from datetime import datetime
import re

# Clean function to fix spacing
def fix_date_spacing(datestr):
    return re.sub(r"([A-Za-z]{3,9})(\d{1,2})", r"\1 \2", datestr)

# Normalize
opening_clean = fix_date_spacing(opening_date)
closing_clean = fix_date_spacing(closing_date)

# Parse
opening_date = datetime.strptime(opening_clean, "%b %d, %Y").date()
closing_date = datetime.strptime(closing_clean, "%b %d, %Y").date()

# Show result
print("Opening Date:", opening_date)
print("Closing Date:", closing_date)
```

```python
import re

# Regex pattern
points_pattern = re.compile(r"(?i)Total Points Earned\s+([\d,]+)")

total_points_earned = None

# Loop through your lines
for line in df.tolist():
    match = points_pattern.search(line)
    if match:
        total_points_earned = match.group(1).replace(",", "")  # remove commas
        break  # Stop at first match

print("Total Points Earned:", total_points_earned)
```

```python
import re

# Lists for results
transaction_dates = []
posting_dates = []
vendors = []
amounts = []

payment_dates = []
payment_posting_dates = []
payment_descriptions = []
payment_amounts = []

# Regex patterns
transaction_pattern = re.compile(
    r"^[^A-Za-z]*"
    r"(?P<transaction_date>[A-Za-z]{3}\s?\d{1,2})\s+"
    r"(?P<posting_date>[A-Za-z]{3}\s?\d{1,2})\s+"
    r"(?P<vendor>.+)$"
)

amount_pattern = re.compile(r"^-?\d{1,4}(?:,\d{3})*(\.\d{1,2})?$")

# Rows
rows = list(enumerate(df.tolist()))

# Temporary storage
transactions = []
payments_with_dates = []
amount_entries = []

# Pass 1: collect transactions and payments
for idx, line in rows:
    line_clean = line.strip()
    match = transaction_pattern.match(line_clean)
    if match:
        vendor_text = match.group("vendor").strip().upper()
        if "PAYMENT RECEIVED" in vendor_text:
            payments_with_dates.append({
                "idx": idx,
                "transaction_date": match.group("transaction_date"),
                "posting_date": match.group("posting_date"),
                "description": match.group("vendor").strip()
            })
        else:
            transactions.append({
                "idx": idx,
                "transaction_date": match.group("transaction_date"),
                "posting_date": match.group("posting_date"),
                "vendor": match.group("vendor").strip()
            })

# Pass 2: collect all amounts
for idx, line in rows:
    line_clean = line.strip()
    if amount_pattern.match(line_clean):
        amt = line_clean.replace(",", "")
        amount_entries.append({
            "idx": idx,
            "amount": amt
        })

# Pair payments with date lines
num_payments_with_dates = len(payments_with_dates)
payment_amounts_list = amount_entries[:num_payments_with_dates]
transaction_amounts_list = amount_entries[num_payments_with_dates:]

# Build payments list
for p, a in zip(payments_with_dates, payment_amounts_list):
    payment_dates.append(p["transaction_date"])
    payment_posting_dates.append(p["posting_date"])
    payment_descriptions.append(p["description"])
    payment_amounts.append(a["amount"])

# Pair transactions
for t, a in zip(transactions, transaction_amounts_list):
    transaction_dates.append(t["transaction_date"])
    posting_dates.append(t["posting_date"])
    vendors.append(t["vendor"])
    amounts.append(a["amount"])

# DataFrames
transactions_df = pd.DataFrame({
    "transaction_date": transaction_dates,
    "posting_date": posting_dates,
    "vendor": vendors,
    "amount": amounts
})

payments_df = pd.DataFrame({
    "transaction_date": payment_dates,
    "posting_date": payment_posting_dates,
    "vendor": payment_descriptions,
    "amount": payment_amounts
})

# Show results
print("✅ Transactions parsed:")
transactions_df

```

```python
# First, split vendor into 2 parts
split_vendors = transactions_df["vendor"].str.rsplit(n=1, expand=True)

# Assign back
transactions_df["vendor"] = split_vendors[0]
transactions_df["location"] = split_vendors[1]

# Show result
transactions_df[["vendor", "location"]]
```

```python
from datetime import datetime
import re


# 1️⃣ Get year from closing date
year = closing_date.year

# 2️⃣ Function to normalize and parse
def parse_partial_date(datestr, year):
    if pd.isna(datestr):
        return None
    # Fix spacing
    datestr = re.sub(r"([A-Za-z]{3,9})(\d{1,2})", r"\1 \2", datestr)
    # Append year
    full_date_str = f"{datestr}, {year}"
    # Parse
    return datetime.strptime(full_date_str, "%b %d, %Y").date()

# 3️⃣ Apply to both columns
transactions_df["transaction_date"] = transactions_df["transaction_date"].apply(
    lambda x: parse_partial_date(x, year)
)

transactions_df["posting_date"] = transactions_df["posting_date"].apply(
    lambda x: parse_partial_date(x, year)
)

# 4️⃣ View results
transactions_df
```
