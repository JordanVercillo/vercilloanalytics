```python

flat_columns = {
    "product_name": "product_name",
    "brands": "brands",
    "code": "barcode",            # <== Flip these
    "categories": "categories",
    "countries_tags": "countries",
    "ingredients_text": "ingredients_text",
    "nutriscore_grade": "nutrition_grade",
    "nova_group": "nova_group",
    "ecoscore_grade": "ecoscore_grade",
    "labels": "labels",
    "allergens_tags": "allergens",
    "packaging": "packaging",
    "created_t": "created_t",
    "last_modified_t": "last_modified_t"
}

# Select and rename
df_flat = df_sample[list(flat_columns.keys())].rename(columns=flat_columns)

# Add image_url as NaN (or extract as shown above)
df_flat["image_url"] = pd.NA
df_flat
```

```python
import numpy as np

def extract_main_text(entry):
    if isinstance(entry, (list, np.ndarray)):
        for d in entry:
            if d.get("lang") == "main":
                return d.get("text")
    return pd.NA

def extract_lang(entry):

    if isinstance(entry, (list, np.ndarray)):
        for d in entry:
            lang = d.get("lang")
            if lang and lang != "main":
                return lang
    return pd.NA

df_flat["product_name_lang"] = df_flat["product_name"].apply(extract_lang)

df_flat['product_name'] = df_flat["product_name"].apply(extract_main_text)
df_flat['ingredients_text']= df_flat["ingredients_text"].apply(extract_main_text)

df_flat["allergens"] = df_flat["allergens"].apply(lambda x: ", ".join(x) if isinstance(x, np.ndarray) else pd.NA)
df_flat["countries"] = df_flat["countries"].apply(lambda x: ", ".join(x) if isinstance(x, np.ndarray) else pd.NA)
```

```python
df_flat.info()
```

```python
# Flatten nutriments
def flatten_nutriments(nutriment_list):
    result = {}
    for item in nutriment_list:
        name = item.get("name")
        value_100g = item.get("100g")
        result[name] = value_100g
    return result

nutriments_flat = df_sample["nutriments"].apply(flatten_nutriments)
nutriments_df = pd.DataFrame(nutriments_flat.tolist())

nutriments_df = nutriments_df.rename(columns={
    "saturated-fat": "saturated_fat_100g",
    "fat": "fat_100g",
    "carbohydrates": "carbohydrates_100g",
    "sugars": "sugars_100g",
    "fiber": "fiber_100g",
    "proteins": "proteins_100g",
    "salt": "salt_100g",
    "energy-kcal": "energy_kcal_100g"
})

nutrients_order = [
    "energy_kcal_100g",
    "fat_100g",
    "saturated_fat_100g",
    "carbohydrates_100g",
    "sugars_100g",
    "fiber_100g",
    "proteins_100g",
    "salt_100g"
]
nutriments_df = nutriments_df.reindex(columns=nutrients_order)
nutriments_df
```

```python
# Concatenate df_flat and nutriments_df horizontally
export_sample = pd.concat([df_flat.reset_index(drop=True), nutriments_df.reset_index(drop=True)], axis=1)

# Reorder columns if you want a specific order
final_column_order = [
    "product_name",
    "product_name_lang",
    "brands",
    "barcode",
    "categories",
    "countries",
    "image_url",
    "ingredients_text",
    "energy_kcal_100g",
    "fat_100g",
    "saturated_fat_100g",
    "carbohydrates_100g",
    "sugars_100g",
    "fiber_100g",
    "proteins_100g",
    "salt_100g",
    "nutrition_grade",
    "nova_group",
    "ecoscore_grade",
    "labels",
    "allergens",
    "packaging",
    "created_t",
    "last_modified_t"
]

export_sample = export_sample.reindex(columns=final_column_order)
export_sample.info()
```
